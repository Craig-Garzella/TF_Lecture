## Second, our dataset 

```
age = np.asarray(adult.age) #putting age values as an array
```

```
income = np.asarray(adult.incomeUSD) #putting income values in an array
```

##  now we use TF!

```
session = tf.Session()
```

```
learning_rate = 0.001 # learning rate specifies how fast and how much data is used
training_epochs = 100 # how many iterations would we want
display_step = 5 #at which steps do we print the output
```

```
train_X = age #X is our input variable
train_y = income #trying to predict
n_samples = train_X.shape[0] # getting the shape of X to know how many samples
```

```
X = tf.placeholder(dtype =tf.float32) # adding placeholders to be replaced each epoch
Y = tf.placeholder(dtype =tf.float32)
```

```
W = tf.Variable(rng.randn(), name = "Weight") #place holders for bias and weight
b = tf.Variable(rng.randn(), name="bias")
```

```
pred = tf.add(tf.multiply(X,W), b) # simple linear regression equation
```

```
mse = tf.reduce_sum(tf.pow(pred-Y, 2)) / (2*n_samples) #calculating the MSE
```

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(mse) 
#using gradient descent to optimize our training
```

```
init = tf.global_variables_initializer() #starting TF env
```

```
with session as sess: 
    sess.run(init)
    for epoch in range(training_epochs):
        for (x,y) in zip(train_X, train_y):
            sess.run(optimizer, feed_dict={X: x, Y: y})
        if (epoch+1) % display_step ==0:
            c = sess.run(mse, feed_dict={X: train_X, Y: train_y})
            print ("Epoch:", '%04d' % (epoch+1), "mse=", "{:.9f}".format(c), \
                "W=", sess.run(W), "b=", sess.run(b))

    print ("Optimization Finished!")
    training_cost = sess.run(mse, feed_dict={X: train_X, Y: train_y})
    print ("Training mse=", mse, "W=", sess.run(W), "b=", sess.run(b), '\n')
      #Graphic display
    plt.plot(train_X, train_y, 'ro', label='Original data')
    plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
    plt.legend()
    plt.show() 
```

```
sess.close()
```
